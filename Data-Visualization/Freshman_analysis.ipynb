## (4) 전국 대학교 전년도 기준 입학생 증감소 비교
- 2019년도 입학자 수 - 2020년도 입학자 수

=> 0보다 클 경우 : 2019 ~ 2020년도 입학자 수 감소

=> 0보다 작을 경우 : 2019 ~ 2020년도 입학자 수 증가

```
freshman_2019.reset_index(inplace = True)
```

```
print(len(freshman_2019['입학자']))
print(len(freshman_2020['입학자']))
# 2019년도와 2020년도 학교수가 달라 nan값 발생 => nan값 제거

univ_name = []
freshman_19to20 =[]

for n in range(len(freshman_2019['학교명'])):
  for i in range(len(freshman_2020['학교명'])):
    if freshman_2019['학교명'][n] == freshman_2020['학교명'][i]:
      univ_name.append(freshman_2020['학교명'][i])
      freshman_19to20.append(freshman_2019['입학자'][n] - freshman_2020['입학자'][i])
    else :
      pass

print(freshman_19to20)
print(univ_name)
```

```
freshman_df = pd.DataFrame({'Name':univ_name, 'cal':freshman_19to20})
```

```
# cal = 2019년도 입학생 수 - 2020년도 입학생 수
freshman_df
```

```
dec_univ = freshman_df[freshman_df['cal']> 0].reset_index()
inc_univ = freshman_df[freshman_df['cal'] <= 0].reset_index()
```

```
dec_univ_lat = []
dec_univ_lng = []

for i, name in enumerate(dec_univ['Name']):
  tmp = gmaps.geocode(name, language = 'ko')
  tmp_loc = tmp[0].get("geometry")
  dec_univ_lat.append(tmp_loc['location']['lat']) # 경도
  dec_univ_lng.append(tmp_loc['location']['lng']) # 위도

  print(name)
  print(dec_univ_lat[i])
  print(dec_univ_lng[i])
```

```
inc_univ_lat = []
inc_univ_lng = []

for i, name in enumerate(inc_univ['Name']):
  tmp = gmaps.geocode(name, language = 'ko')
  tmp_loc = tmp[0].get("geometry")
  inc_univ_lat.append(tmp_loc['location']['lat']) # 경도
  inc_univ_lng.append(tmp_loc['location']['lng']) # 위도

  print(name)
  print(inc_univ_lat[i])
  print(inc_univ_lng[i])
```

```
map2 = folium.Map(location = [36.02, 127.01], zoom_start = 8)

# 감소 
for n in range(len(dec_univ['Name'])):
  folium.CircleMarker([dec_univ_lat[n], dec_univ_lng[n]],
                      radius = dec_univ['cal'][n]*0.1,
                      color = '#3186cc', fill_color = '#3186cc').add_to(map2)

# 증가
for n in range(len(inc_univ['Name'])):
  folium.CircleMarker([inc_univ_lat[n], inc_univ_lng[n]],
                      radius = inc_univ['cal'][n]*(-0.1),
                      color = 'red', fill_color = 'red').add_to(map2)
map2
```
